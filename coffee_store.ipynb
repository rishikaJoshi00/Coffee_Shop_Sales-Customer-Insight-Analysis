```python
import sqlite3

# Create connection
conn = sqlite3.connect('sales.db')

# Save dataframe into SQL table
df.to_sql('sales', conn, if_exists='replace', index=False)

print("✅ SQL table created!")
```

```python
query = """
SELECT product_category, SUM(total_amount) AS total_sales
FROM sales
GROUP BY product_category
ORDER BY total_sales DESC;
"""

result = pd.read_sql(query, conn)
result
```

```python
query = """
SELECT product_name, COUNT(DISTINCT customer_id) AS num_customers
FROM sales
GROUP BY product_name
ORDER BY num_customers DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT product_name, SUM(quantity) AS total_quantity
FROM sales
GROUP BY product_name
ORDER BY total_quantity DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT loyalty_member,
       COUNT(*) AS transactions,
       SUM(total_amount) AS total_spent,
       AVG(total_amount) AS avg_spent
FROM sales
GROUP BY loyalty_member;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT city, SUM(total_amount) AS total_sales
FROM sales
GROUP BY city
ORDER BY total_sales DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT weather_condition, AVG(total_amount) AS avg_sales
FROM sales
GROUP BY weather_condition;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT product_category,
       SUM(total_amount) AS revenue,
       SUM(quantity) AS total_units,
       COUNT(DISTINCT customer_id) AS num_customers
FROM sales
GROUP BY product_category
ORDER BY revenue DESC;
"""

pd.read_sql(query, conn)
```
